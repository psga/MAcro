# Funciones 

**Funciones de una Variable real:**

$ x \rightarrow f(x)$

o   donde $x\in [a.b)$, es decir con  $a\leq x < b$

supongamos  $-20\leq x < 20$

```python
#generamos un vector con valores de x en [-20,20] , 
#no podemos tomar toooodos los valores como si fuera continuo, 
#tomemos los ms posibles por ej n=100 o n=10000
n=100 # cantidad de muestras 
a=-20 #inicio
b=20
x=np.linspace(a,b,n, endpoint=False)
print(x)
print('cantidad de muestras de x:',n)
```

Si $f(x)= sen(x)$ 

la derivada de $f$ es

$f'(x)=\frac{f(x)}{dx}= cos(x)$ .Las grafiquemos para $-\pi\leq x < \pi$


```python
n=400 # cantidad de muestras 
(a,b)=(-np.pi ,np.pi)
x=np.linspace(a,b,n, endpoint=False)
f=np.sin(x) # vector, contiene la funcin evaluada p/c valor de x 
f_1=np.cos(x)
```

```python
#Visualizamos con plot y subplot
plt.figure(figsize=(16,10))
plt.subplot(211)
plt.plot(x,f, label='sen(x)')
plt.xlabel('x (muestreado)')
plt.ylabel('f(x)=sen(x)')
plt.grid()
plt.axvline(-np.pi/2, color = "red", linewidth = 1, linestyle = "dashed")
plt.axvline(0, color = "red", linewidth = 1, linestyle = "dashed");
plt.axvline(np.pi/2, color = "red", linewidth = 1, linestyle = "dashed");
plt.legend()
plt.subplot(212)
plt.plot(x,f_1, label='cos(x)')
plt.xlabel('x (muestreado)')
plt.ylabel('$f^{-1}(x)=cos(x)$')
plt.grid()
plt.axvline(-np.pi/2, color = "red", linewidth = 1, linestyle = "dashed")
plt.axvline(0, color = "red", linewidth = 1, linestyle = "dashed");
plt.axvline(np.pi/2, color = "red", linewidth = 1, linestyle = "dashed");
plt.legend(loc='upper right')

plt.show()
```

Si $f(x)= x^2$ 

la derivada de $f$ es

$f'(x)=\frac{d f(x)}{dx}= 2x$ .

la primitiva (integral indefinida) de $f$ es

$g(x)=\frac{1}{3}x^3$, tal que $g'(x)=f(x)$.


Las grafiquemos para $-2\leq x < 2$


```python
n=400 # cantidad de muestras 
(a,b)=(-3 ,3)
x=np.linspace(a,b,n, endpoint=False)
f=x**2 # vector, contiene la funcin evaluada p/c valor de x 
df=2*x #derivada de f
af=x**3/3 #integral (area bajola curva) de f
```

```python
#Visualizamos con plot y subplot
plt.figure(figsize=(16,10))
plt.subplot(311)
plt.plot(x,f, label='x^2')
plt.xlabel('x ')
plt.ylabel('$f(x)=x^2$')
plt.grid()
plt.legend()

plt.subplot(312)
plt.plot(x,df, label='derivada de f=2x')
plt.xlabel('x')
plt.ylabel('derivada de f=2x')
plt.grid()
plt.axvline(0, color = "red", linewidth = 1, linestyle = "dashed");
plt.legend()

plt.subplot(313)
plt.plot(x,af, label='integral de f: $x^3/3$')
plt.xlabel('x')
plt.ylabel('integral de f: $x^3/3$')
plt.grid()
plt.axvline(0, color = "red", linewidth = 1, linestyle = "dashed");
plt.legend()

plt.show()
```

#  Varias variables.

* Varias variables. Derivadas parciales.

$f(x,y)$ con $x$ e $y$ nmeros reales

Por ejemplo:

$f(x,y)=sen(x).cos(y)$

Las derivadas parciales de f son:

$\frac{\partial f(x,y)}{\partial x}= cos(x).cos(y)$ .

$\frac{\partial f(x,y)}{\partial y}= - sen(x).sen(y)$ .


```python
n=400 
(a,b)=(-np.pi ,np.pi)
x=np.linspace(a,b,n, endpoint=False)
y=np.linspace(a,b,n, endpoint=False)
f=np.zeros((n,n))
for i in range(n):
  f[i]= np.sin(x[i])*np.cos(y)

```

```python
#grfico 3D
from mpl_toolkits.mplot3d import axes3d
# Creamos la figura
fig = plt.figure()
# Creamos el plano 3D
ax1 = fig.add_subplot(111, projection='3d')
#discretizamos en k=5 a x para (k pueden ser ms o menos) un color para cada x 
k=5 
paso=n//k
for i in range(k):
  xn=np.full(n,x[80*i])
  ax1.plot3D(xn,y,f[80*i], label='x='+str(x[80*i]))
plt.show()
```
